```python
#But what we did was just getting the authors from the first page but now what if I want to get authors from all th first 10 pages 
```

```python
base_url = "https://quotes.toscrape.com/page/{}/"
```

```python
base_url.format(1)    #Perfect!!!
```

```python
for n in range(11):
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    soup = bs4.BeautifulSoup(res.text , "lxml")
    authors = soup.select(".author")
    
    for author in authors:
        print(author.getText())
        
```

```python
#let's see if we can get a unique list of all the authors
```

```python
for n in range(11):
    my_authors = []
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    soup = bs4.BeautifulSoup(res.text , "lxml")
    authors = soup.select(".author")
    
    for author in authors:
         my_authors.append(author.getText())
        
```

```python
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
```

```python
base_url.format(3)  #perfect
```

```python
two_star_titles = []

for n in range(51):
    
    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(res.text , "lxml")
    books = soup.select(".product_pod")
    
    for book in books:
        
        if len(book.select(".star-rating.Two")) != 0:
            book_title = book.select("a")[1]["title"]
            two_star_titles.append(book_title)
```
